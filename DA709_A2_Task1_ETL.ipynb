```python
#Indentify missing values
df.isnull().sum()
```

```python
#Identify if there is duplicate values in the column "id"
duplicate_ids = df[df['id'].duplicated()]
```

```python
duplicate_ids.shape
```

```python
duplicate_ids.head()
```

```python
# Remove duplicate rows based on the 'id' column, keeping the first occurrence
df = df.drop_duplicates(subset='id', keep='first')

# Display the number of remaining rows and a preview of the dataset
print(f"Number of rows after removing duplicates: {df.shape[0]}")
df.head()
```

```python
# save the cleaned dataset to a csv file
df.to_csv("cleaned_spotify_tracks.csv", index=False)
```
